```python
import pandas as pd
```

```python
df_path = "api_data.csv"
```

```python
df = pd.read_csv(df_path)
```

```python
df.info()
```

```python
df
```

```python
del df["Awards"]
```

```python
del df["Ratings"]
```

```python
del df["Unnamed: 0"]
```

```python
reduced = df.dropna()
```

```python
reduced.info()
```

```python
reduced['Runtime'] = reduced['Runtime'].str.extract('(\d+)').astype(int)
reduced
```

```python
reduced['Box_Office'] = reduced['Box_Office'].str.replace("$", "")
reduced
```

```python
reduced['Box_Office'] = reduced['Box_Office'].str.replace(",", "")
reduced
```

```python
reduced.info()
```

```python
reduced['Box_Office'] = reduced['Box_Office'].astype("int")
reduced
```

```python
reduced.info()
```

```python
reduced.to_csv("reduced_api_data.csv")
```
